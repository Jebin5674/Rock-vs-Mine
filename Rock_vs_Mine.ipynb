```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data = pd.read_csv('/content/sonar data.csv', header = None)
```

```python
sonar_data.head()
```

```python
sonar_data.describe()
```

```python
sonar_data.shape
```

More the data more accurate tha data will be


```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

Seperating tha datas in "X" and output in "Y"

```python
X=sonar_data.drop(columns=60,axis = 1)
Y=sonar_data[60]
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model = LogisticRegression()
```

```python
print(X_train)
print(Y_train)
```

```python
model.fit(X_train, Y_train)
```

```python
#accuracy for tranining
X_train_prediction = model.predict(X_train)
traning_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy for traning",traning_data_accuracy)
```

```python
#accuracy for test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy for testing", test_data_accuracy)
```
